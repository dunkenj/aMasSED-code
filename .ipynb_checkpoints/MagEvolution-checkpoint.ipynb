## Import required modules and set up cosmology

```python
%matplotlib inline
import seaborn as sns
import smpy as S
import numpy as np
import matplotlib.pyplot as plt

import astropy.units as u
from astropy.cosmology import FlatLambdaCDM

cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
sns.set_style('white')
sns.set_context('talk')
```

## Choose stellar population parameters

#### Initial formation redshift:

```python
z_formation = 4.
```

#### Redshift range to observe models at and corresponding ages:

```python
zrange = np.linspace(0., 3, 30)
sfh_taus = np.array([0.5,1.,2.5,5.]) * u.Gyr # star-formation history tau's

ages = cosmo.age(zrange) - cosmo.age(z_formation) # Galaxy age
distmod = cosmo.distmod(zrange) # Corresponding distance moduli

print('Ages range from {0:.1f} to {1:.1f}'.format(ages[-1], ages[0]))
```

## Build model SEDs

```python
path = '/Users/ken/Documents/PhD/code/bc03/models/Padova1994/chabrier/bc2003_lr_BaSeL_m'
ssp = S.BC(path)
gal = S.CSP(ssp)

gal.build(ages, sfh_taus, 0., 1., 1.) # Make galaxy SEDs for each age and SFH
                                      # assuming 0. dust, solar metallicity, and fesc=1
```

### Convolve through H band filter at each wavelength

```python
Filts = S.FilterSet('fitting/GOODS-S_18_FilterCurves/Filter10_WFC3_f160w.txt') # Load H filter
ModelFluxes = S.Observe(gal, Filts, zrange, force_age=False) # Convolve models with H at all redshifts

magnitudes = np.squeeze(ModelFluxes.AB) * u.mag # Remove unnecessary dimensions
```

## Plots

```python
Fig, Ax = plt.subplots(figsize=(7,4.5))

# Since the ages correspond 1:1 with the observed redshifts, 
# we want the diagonal the 2D array of magnitudes for each 
# SFH.
for i, tau in enumerate(sfh_taus):
    Ax.plot(zrange, distmod-magnitudes[:,:,i].diagonal(),
            lw = 2,
            label='SFH = {0:.1f}'.format(tau))
    
Legend = Ax.legend(loc='upper left')
Ax.set_xlim(zrange.min(), zrange.max())
Ax.set_xlabel('Redshift, z')
Ax.set_ylabel(r'Distance modulus - $H_{160}(z)$') # Subtracted from dist. modulus to seperate the stellar population mag change
```

```python
Fig, Ax = plt.subplots(figsize=(7,4.5))
for i, tau in enumerate(sfh_taus):
    Ax.plot(zrange, magnitudes[:,:,i].diagonal(), 
            lw = 2,
            label='SFH = {0:.1f}'.format(tau))
    
Legend = Ax.legend(loc='lower right')
Ax.set_xlim(zrange.min(), zrange.max())
Ax.set_ylim([46,49])
Ax.set_xlabel('Redshift, z')
Ax.set_ylabel(r'$H_{160}(z)$')
```
