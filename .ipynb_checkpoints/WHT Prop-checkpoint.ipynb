```python
%matplotlib inline
import numpy as np
import astropy.units as u
import smpy as S
import matplotlib.pyplot as plt

EAZYfilters = S.LoadEAZYFilters('FILTER.RES.V8')
```

```python
path = '/Users/ken/Documents/Astro/code/bc03/models/Padova1994/chabrier/bc2003_lr_BaSeL_m'
ssp = S.BC(path)
lowz = S.CSP(ssp)

lowz.build(0.1*u.Gyr, 5*u.Gyr, 0., 0.5, 0.) # Make galaxy SEDs for each age and SFH
                                      # assuming 0. dust, solar metallicity, and fesc=1

highz = S.CSP(ssp)
highz.build(0.1*u.Gyr, 5*u.Gyr, 0, 1., 0.)
highz.addEmissionLine(1216*u.AA, 100*u.AA)
```

```python
Filts = S.FilterSet()
x = zip(np.arange(len(EAZYfilters.filternames)), EAZYfilters.filternames)
Filts.addEAZYFilter(EAZYfilters,[140,142,143,82,125,126])
```

```python
z1 = 0.75
plt.plot(lowz.wave*(1+z1),np.squeeze(lowz.SED)*10, label='z = {0}'.format(z1))

z2 = 6.25
plt.plot(highz.wave*(1+z2),np.squeeze(highz.SED),label='z = {0}'.format(z2))
# Illustrate filter responses
for filt in Filts.filters:
    plt.fill_between(filt.wave,(filt.response/np.max(filt.response))*0.2, alpha=0.5, color='gray')

leg = plt.legend(loc='upper left')
#plt.yscale('log')
plt.xlim(3000,14000)
plt.ylim(0, 0.5)
```

```python
zr1 = np.linspace(0,2,30)
lowz_Mags = S.Observe(lowz, Filts, zr1)

zr2 = np.linspace(5,7,30)
hiz_Mags = S.Observe(highz, Filts, zr2)
```
