## Import required modules and set up cosmology

```python
%matplotlib inline
import seaborn as sns
import smpy as S
import numpy as np
import matplotlib.pyplot as plt

import astropy.units as u
from astropy.cosmology import FlatLambdaCDM

cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
sns.set_style('white')
sns.set_context('talk')

EAZYfilters = S.LoadEAZYFilters('FILTER.RES.V8')
```

## Calculating A(V) to A(NUV) conversion

```python
def k_lambda_calz(wave):
    k = np.zeros_like(wave.value)

    w0 = [wave <= 1200 * u.AA]
    w1 = [wave < 6300 * u.AA]
    w2 = [wave >= 6300 * u.AA]
    w_u = wave.to(u.um).value

    x1 = np.argmin(np.abs(wave - 1200 * u.AA))
    x2 = np.argmin(np.abs(wave - 1250 * u.AA))

    k[w2] = 2.659 * (-1.857 + 1.040 / w_u[w2])
    k[w1] = 2.659 * (-2.156 + (1.509 / w_u[w1]) - (0.198 / w_u[w1] ** 2) + (0.011 / w_u[w1] ** 3))
    k[w0] = k[x1] + ((wave[w0] - 1200. * u.AA) * (k[x1] - k[x2]) / (wave[x1] - wave[x2]))

    k += 4.05
    k[k < 0.] = 0.
    return k

FiltsDust = S.FilterSet()
FiltsDust.addEAZYFilter(EAZYfilters,[120,136]) 
NUV = FiltsDust.filters[0] # GALEX NUV
Vband = FiltsDust.filters[1] # Bessel V band

kNUV = np.trapz(k_lambda_calz(NUV.wave)*NUV.response, NUV.wave) / np.trapz(NUV.response, NUV.wave)
kV = np.trapz(k_lambda_calz(Vband.wave)*Vband.response, Vband.wave) / np.trapz(Vband.response, Vband.wave)

print('Filter: {0}, Central Wavelength: {1}, k: {2}'.format('Galex NUV',NUV.lambda_c, kNUV))
print('Filter: {0}, Central Wavelength: {1}, k: {2}'.format('V',Vband.lambda_c, kV))
print kNUV/kV
# A(NUV) = (k(NUV)/k(V)) * A(V)
# A(V) = k(V)/k(NUV) * A(NUV)
```

## Make colour tracks and plot

```python
# Set up plot, filters, SSP models
Fig, Ax = plt.subplots(figsize=(7,4.5))
Filts = S.FilterSet('filters/filt_0*.txt') # Load H filter

path = '/Users/ken/Documents/PhD/code/bc03/models/Padova1994/chabrier/bc2003_lr_BaSeL_m'
ssp = S.BC(path)

# Plotting styles
ls = ['-', '--', ':']
cols = ['steelblue', 'firebrick', 'olivedrab']

# SED/Redshift properties
z_formation = np.array([3., 4., 5.])
zrange = np.linspace(0., 3, 40)

sfh_taus = np.array([0.1]) * u.Gyr # star-formation history tau's
dusts = np.array([0, 1., 2.]) # Desired A(NUV) extinction, converted to Av below

for iz, zform in enumerate(z_formation):
    ages = cosmo.age(zrange) - cosmo.age(zform) # Galaxy age

    #print('Ages range from {0:.1f} to {1:.1f}'.format(ages[-1], ages[0]))
    gal = S.CSP(ssp)
    gal.build(ages, sfh_taus, dusts * kV/kNUV, 1.0, 1.) # Make galaxy SEDs for each age and SFH

    # Convolve with filters and get colours
    ModelFluxes = S.Observe(gal, Filts, 0., force_age=False)
    magnitudes = np.squeeze(ModelFluxes.AB) * u.mag # Remove unnecessary dimensions
    U_V = magnitudes[0]-magnitudes[1]

    # Plot lines
    for i, dust in enumerate(dusts):
        Ax.plot(zrange, U_V[:,i], color=cols[i],
                linestyle=ls[iz],
                lw = 2,
                label=r'{0} = {1:.0f}, {2} = {3:.1f}'.format('$z_{f}$', zform, '$A_{NUV}$', dust))

Legend = Ax.legend(loc='upper right', ncol=3, prop={'size':9})
Ax.set_xlim(zrange.min(), zrange.max())
Ax.set_ylim([0.5,3.5])
Ax.set_xlabel('Redshift, z')
Ax.set_ylabel(r'U-V')

Fig.savefig('UV_z_tracks_1.0Zsol.pdf',format='pdf')
```
