Import required modules and set up cosmology

```python
%matplotlib inline
import seaborn as sns
import smpy as S
import ssp as B
import numpy as np
import matplotlib.pyplot as plt

import astropy.units as u
from astropy.cosmology import FlatLambdaCDM
from astropy.table import Table
from astropy.utils.console import ProgressBar

cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
sns.set_style('white')
sns.set_style('ticks')
sns.set_context('talk')

EAZYfilters = S.LoadEAZYFilters('FILTER.RES.V8')
```

Load in Pentericci+ 2001 HzRG sample:

```python
pent = Table.read('pentericci_table.tex',format='ascii.commented_header')
```

Now building a couple of normalised galaxy templates to scale to the observations at their observed redshifts and apparent magnitudes.

One is a very blue galaxy (with minimal dust and low metallicity) and the other is significantly redder, with 1 Av of dust and solar metallicity. Both are ~constant SFR galaxies at 1 Gyr old and nebular emission lines included (no Lyman alpha).

```python
path = '/Users/ken/Documents/Astro/code/bc03/models/Padova1994/salpeter/bc2003_lr_BaSeL_m'
BC03 = B.BC(path)

blue = S.CSP(BC03)
dusty = S.CSP(BC03)
```

For each of the galaxies in the Pentericci HzRG sample, now observing both galaxy models at the HzRG galaxy's spectroscopic redshift and finding the normalisation (~stellar mass effectively for these models) which matches the observed NICMOS H band magnitude.

The relevant normalisation (and the star-formation rate which corresponds) for each galaxy are then saved as an extra column.

```python
NICMOS = S.FilterSet()
NICMOS.addEAZYFilter(EAZYfilters, [8])

z_formation = 9.

pent['scale'] = np.zeros(len(pent))
pent['sfr'] = np.zeros(len(pent))

pent['scale_dusty'] = np.zeros(len(pent))
pent['sfr_dusty'] = np.zeros(len(pent))

with ProgressBar(len(pent)) as bar:
    for i, row in enumerate(pent):
        # Set stellar pop. age as time since z_formation
        age = cosmo.age(row['z'])-cosmo.age(z_formation)

        blue.build(age, 100*u.Gyr, 0.1, 0.2, fesc=0.)
        obs = S.Observe(blue, NICMOS, row['z'])
        scale = 10**((np.squeeze(obs.AB.value)-row['mag'])/2.5)
        pent[i]['scale'] = np.log10(scale)
        pent[i]['sfr'] = np.squeeze(blue.SFR.value) * scale

        dusty.build(age, 100*u.Gyr, 1., 1., fesc=0.)
        obs = S.Observe(dusty, NICMOS, row['z'])
        scale = 10**((np.squeeze(obs.AB.value)-row['mag'])/2.5)
        pent[i]['scale_dusty'] = np.log10(scale)
        pent[i]['sfr_dusty'] = np.squeeze(dusty.SFR.value) * scale
        bar.update()
```




Now that I know the normalisation for each model to match the observed HzRG photometry, can then reobserve said templates at higher redshift to see what apparent magnitudes they would be in any of the PanSTARRS (or other) filters.

```python
Fig, Ax = plt.subplots(2,3,figsize=(16,8))
Ax = Ax.flatten()
grizy = S.FilterSet('filters/PanStarrs_0*.txt')

PS_md_depths = [25.8, 25.7, 25.9, 25.4, 24.]
PS_ps1_depths = [23.35, 22.75, 22.55, 21.8, 21.25] # From george
bnames = ['g', 'r', 'i', 'z', 'y']

Legends = []

for band in np.arange(5):
    print(bnames[band])
    for i, row in enumerate(pent):
        zrange = np.linspace(row['z'], 8., 30)

        ages = cosmo.age(zrange)-cosmo.age(z_formation)
        max_age = cosmo.age(row['z'])

        # Bluer template
        scales = 10**row['scale']  * (ages/max_age)
        gal = blue / blue.Ms
        gal.build(ages, 100*u.Gyr, 0.1, 0.2, fesc=0.)
        obs = S.Observe(gal , grizy, zrange)

        # np.diagonal matches redshift array with corresponding ages
        mags = np.diagonal(np.squeeze(obs.AB[:,band])) - 2.5*np.log10(scales)
        Ax[band].plot(zrange, mags,color='steelblue', alpha=0.2)

        # Redder template
        dusty_scales = 10**row['scale_dusty'] * (ages/max_age)
        gal = dusty / dusty.Ms
        gal.build(ages, 100*u.Gyr, 1., 1., fesc=0.)
        obs = S.Observe(gal , grizy, zrange)

        mags = np.diagonal(np.squeeze(obs.AB[:,band])) - 2.5*np.log10(dusty_scales)
        Ax[band].plot(zrange, mags, color='firebrick', alpha=0.2)


# Show PanSTARRS depths
    Ax[band].plot([0,10],[PS_ps1_depths[band],PS_ps1_depths[band]], '-.', lw=3, color='darkgreen',
                  label='Approx. 5$\sigma$ PS1 Depth')
    #Ax[band].text(2., PS_ps1_depths[band]+0.2, 'Approx. 5$\sigma$ PS1 Depth',
    #        verticalalignment='top',
    #        color='darkgreen',size=12)

    Ax[band].plot([0,10],[PS_md_depths[band],PS_md_depths[band]], '--', lw=3, color='forestgreen',
                  label = r'Approx. 5$\sigma$ MD Depth')
    #Ax[band].text(2., PS_md_depths[band]+0.2, r'Approx. 5$\sigma$ MD Depth',
    #        verticalalignment='top',
    #        color='forestgreen',size=12)
    Leg = Ax[band].legend(loc='upper right')
    Leg.draw_frame(False)
    Legends.append(Leg)
    Ax[band].set_title('{0:s} band'.format(bnames[band]))
    Ax[band].set_xlim(1.5,7)
    Ax[band].set_ylim(29,17)
    Ax[band].set_xlabel(r'Redshift, $z$')
    Ax[band].set_ylabel('Apparent magnitude')

Ax[-1].axis('off')
Fig.subplots_adjust(hspace=0.32, wspace=0.25, left=0.06)
Fig.savefig('HzRG_observability.pdf',format='pdf')
```
